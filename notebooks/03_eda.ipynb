```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Load the cleaned dataset
df = pd.read_csv('/content/Zomato_Cleaned.csv')

print("Cleaned dataset loaded!")
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

```python
# Average delivery time by city
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='City', y='Time_taken (min)',
            hue='City', legend=False, palette='Reds')
plt.title('Average Delivery Time by City')
plt.xlabel('City')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

```python
# Average delivery time by weather conditions
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Weather_conditions', y='Time_taken (min)',
            hue='Weather_conditions', legend=False, palette='Reds')
plt.title('Average Delivery Time by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

```python
# Average delivery time by weather conditions
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Road_traffic_density', y='Time_taken (min)',
            hue='Road_traffic_density', legend=False,
            order=['Low', 'Medium', 'High', 'Jam'], palette='Reds')
plt.title('Average Delivery Time by Traffic Density')
plt.xlabel('Traffic Density')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

```python
# Average delivery time by road traffic density
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Road_traffic_density', y='Time_taken (min)',
            hue='Road_traffic_density', legend=False,
            order=['Low', 'Medium', 'High', 'Jam'], palette='Reds')
plt.title('Average Delivery Time by Traffic Density')
plt.xlabel('Traffic Density')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

```python
# Festival impact on delivery time
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Festival', y='Time_taken (min)',
            hue='Festival', legend=False, palette='Reds')
plt.title('Impact of Festival on Delivery Time')
plt.xlabel('Festival')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

```python
# Delivery time by vehicle type
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Type_of_vehicle', y='Time_taken (min)',
            hue='Type_of_vehicle', legend=False, palette='Reds')
plt.title('Average Delivery Time by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

```python
# Ratings vs delivery time
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Delivery_person_Ratings', y='Time_taken (min)',
                color='red', alpha=0.3)
plt.title('Delivery Person Ratings vs Delivery Time')
plt.xlabel('Ratings')
plt.ylabel('Time Taken (min)')
plt.show()
```

```python
# Orders by hour of day
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='Order_Hour', hue='Order_Hour', legend=False, palette='Reds')
plt.title('Number of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Orders by day of week
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Day_of_Week', hue='Day_of_Week', legend=False, palette='Reds',
              order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Number of Orders by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# Impact of multiple deliveries on delivery time
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='multiple_deliveries', y='Time_taken (min)',
            hue='multiple_deliveries', legend=False, palette='Reds')
plt.title('Impact of Multiple Deliveries on Delivery Time')
plt.xlabel('Number of Multiple Deliveries')
plt.ylabel('Avg Time Taken (min)')
plt.show()
```

## EDA Observations

### Delivery Time
- Average delivery time is around 26 minutes
- Delivery time ranges from 10 to 54 minutes

### City
- Metropolitan cities have higher average delivery times than Urban and Semi-Urban

### Weather
- Sandstorms and Stormy weather cause the highest delivery delays
- Sunny weather has the lowest delivery times

### Traffic
- Jam traffic density causes the highest delivery times
- Low traffic has the lowest delivery times

### Festival
- Festival days have higher delivery times than non-festival days

### Vehicle Type
- Motorcycles and scooters are the most common vehicles
- Bicycles have the highest delivery times

### Order Type
- All order types have similar delivery times

### Ratings
- Higher rated delivery partners tend to deliver slightly faster

### Peak Hours
- Most orders are placed during lunch and dinner hours

### Day of Week
- Weekends have higher order volumes than weekdays

### Multiple Deliveries
- More multiple deliveries = higher delivery time
